```python
model_df = pd.read_csv("Data/transform.csv")
model_df
```

```python
model_df_org = pd.read_csv("Data/transform_org.csv")
model_df_org
```

```python
data_dt = pd.read_csv("Data/Validation.csv")
```

```python
x = model_df_org.iloc[:,0:289] 
y = model_df_org.iloc[:,[289]]
```

<h2><strong>Linear Regression</strong><h2>

```python
#LinearRegression
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=24)
lr = LinearRegression()
lr_fit = lr.fit(x_train,y_train)
```

```python
pre = lr_fit.predict(x_test)
pre
```

```python
r2_score(y_test,pre)
```

```python
x_test.to_csv("Data/X_test.csv",index=False)
y_test.to_csv("Data/y_test.csv",index=False)
```

<h2><strong>Decision Tree</strong><h2>

```python
#DecisionTree
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=24)
dt = DecisionTreeRegressor()
dt_train = dt.fit(x_train,y_train)
dt_train
```

```python
pred = dt_train.predict(x_test)
```

```python
r2_score(y_test,pred)
```

```python
#Tuning dt using RandomizedCV
params = {
    'max_depth': [4, 5, 6, 8, 10, None],
    'min_samples_split': [2, 4, 6, 10],
    'min_samples_leaf': [1, 2, 4, 6],
    'max_features': ['auto', 'sqrt', 'log2', None]
}
random = RandomizedSearchCV(estimator=dt,param_distributions = params,cv=5,n_iter=20,random_state=12,n_jobs=-1,scoring="r2")
tune_dt = random.fit(x_train,y_train)
tune_dt
```

```python
new_dt = DecisionTreeRegressor(min_samples_leaf=6)
new_dt_fit = new_dt.fit(x_train,y_train)
new_dt_fit
```

```python
new_pred = new_dt_fit.predict(x_test)
```

```python
#score
r2_score(y_test,new_pred)
```

<h2><strong>RandomForest</strong><h2>

```python
RandomF = RandomForestRegressor()
R_fit = RandomF.fit(x_train,y_train)
R_fit
```

```python
R_pre = R_fit.predict(x_test)
R_pre
```

```python
r2_score(y_test,R_pre)
```

```python
importance = R_fit.feature_importances_
importance
```
